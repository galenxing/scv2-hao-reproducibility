# Runs QC (filtering of mitochondrial genes + libsize)

```python
import anndata
import scanpy as sc
import numpy as np
import os
```

# QC on round1round2 data

```python
adata = anndata.read('../data/round1round2_raw.h5ad')
```

```python
adata
```

```python
adata.var['mt'] = adata.var_names.str.startswith('MT-')  # annotate the group of mitochondrial genes as 'mt'
sc.pp.calculate_qc_metrics(adata, qc_vars=['mt'], percent_top=None, log1p=False, inplace=True)
```

```python
sc.pl.violin(adata, ['n_genes_by_counts', 'total_counts', 'pct_counts_mt'],
             jitter=0.4, multi_panel=True)
```

```python
sc.pl.scatter(adata, x='total_counts', y='pct_counts_mt')
sc.pl.scatter(adata, x='total_counts', y='n_genes_by_counts')
```

```python
min_libsize = 200
max_libsize = 300000
max_mito_pct = 20

libsize = adata.obs['library_size']

libsize_mask = (min_libsize < libsize) & (libsize < max_libsize)
adata = adata[libsize_mask].copy()

adata = adata[adata.obs.pct_counts_mt < max_mito_pct, :].copy()
```
